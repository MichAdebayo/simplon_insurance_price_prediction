```python
import pandas as pd
import missingno as msno
import seaborn as sns
import matplotlib.pyplot as plt
```

### Partie Analyse :

1. **Vérifiez les informations manquantes et les doublons** (outil : `missingno`).

```python
# Load dataset

df = pd.read_csv("assurance_dataset.csv")
```

```python
# Make a copy of raw dataset

dfi = df.copy()
```

```python
# View first 10 data rows

dfi.head(10)
```

```python
# View data types for each variable

dfi.dtypes
```

```python
# Check for missing data (in numbers)

dfi.isna().sum()
```

```python
# Check for missing data (vizualization)

msno.bar(dfi) 
```

```python
# Check for duplicates

dfi.duplicated().count()
```

```python
# Round bmi to 2 decimal places

dfi['bmi'] = dfi['bmi'].apply(lambda x: round(x, 2))
```

```python
# Round charges to 2 decimal places

dfi['charges'] = dfi['charges'].apply(lambda x: round(x, 2))
```

```python
dfi
```

```python
# Convert smokers column to numeric

dfi['smoker'] = dfi['smoker'].map({'yes': 1, 'no': 0})
```

```python
# Strip sex column data
dfi['sex'] = dfi['sex'].str.strip()
```

```python
# Convert sex column to numeric

dfi['sex'] = dfi['sex'].map({'male': 1, 'female': 0})
```

2. **Détection des valeurs aberrantes.**

```python
# Distribution of the insured by age

sns.displot(dfi, x = 'age', kind='hist')
plt.title("Distribution of the insured by age", fontsize= 13, weight = "bold")
plt.xlabel("Age", weight = 10)
plt.ylabel("Frequency", weight = 10)
```

```python
# Distribution of the insured by bmi (body:mass index)

sns.displot(dfi, x = 'bmi', kind='hist')
plt.title("Distribution of the insured by BMI", fontsize= 13, weight = "bold")
plt.xlabel("BMI", weight = 10)
plt.ylabel("Frequency", weight = 10)
```

```python
# Find number of people with "outlier" BMI values

num_with_high_bmi = dfi.loc[df['bmi'] > 48, :].count()

num_with_high_bmi
```

```python
# Drop individuals with "outlier" BMI values

df_no_outlier_bmi = dfi.loc[dfi['bmi'] <= 48, :]
```

```python
# View latest distribution by bmi

sns.displot(df_no_outlier_bmi, x = 'bmi', kind='hist')
plt.title("Distribution of the insured by BMI", fontsize= 13, weight = "bold")
plt.xlabel("BMI", weight = 10)
plt.ylabel("Frequency", weight = 10)
```

```python
# Distribution of the insured by charges

sns.displot(df_no_outlier_bmi, x = 'charges', kind='hist')
plt.title("Distribution of the insured by charges", fontsize= 13, weight = "bold")
plt.xlabel("Charges (€)", weight = 10)
plt.ylabel("Frequency", weight = 10)
```

```python
# Find number of individuals with high insurance charges

num_with_high_charges = df_no_outlier_bmi.loc[df['charges'] > 50000, :].count()

num_with_high_charges
```

```python
df_no_outlier_charges = df_no_outlier_bmi.loc[df_no_outlier_bmi['charges'] <= 50000, :]

df_no_outlier_charges
```

```python
# # View latest distribution by charges

sns.displot(df_no_outlier_charges, x = 'charges', kind='hist')
plt.title("Distribution of the insured by charges", fontsize= 13, weight = "bold")
plt.xlabel("Charges (€)", weight = 10)
plt.ylabel("Frequency", weight = 10)
```

```python
# Export cleaned data to csv

df_no_outlier_charges.to_csv("cleaned_insurance_data.csv")
```
